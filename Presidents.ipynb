```python
import pandas as pd
import re
```

```python
df = pd.read_csv('presidents.csv',index_col=0)
df
```

```python
def splitname(row):
    row['Firstname']=row['President'].split(' ')[0]
    row['Lastname']=row['President'].split(' ')[-1]
    return row

df=df.apply(splitname,axis='columns')    
df.head()
```

```python
pattern = r'([\w]{3} [\d]{1,2}, [\d]{4})'
df['Born'].str.extract(pattern)
```

```python
df['Born'] = df['Born'].str.extract(pattern)
```

```python
df['Born'] = pd.to_datetime(df['Born'])
df.head()
```

```python
df['Age at the start of Presidency'] = df['Age atstart of presidency'].str.extract(r'(^\d{2}\syears,\s\d{1,3}\sdays)')
df.head()
```

```python
df=df.rename(columns={'Age atstart of presidency':'Start of Presidency'})
df.head()
```

```python
df['Start of Presidency'] = df['Start of Presidency'].str.extract(r'(?:^.*days)(\w{3}\s\d{1,2},\s\d{4})')
df.head()
```

```python
df['Start of Presidency'] = pd.to_datetime(df['Start of Presidency'])
df.head()
```

```python
df['Age at the end of Presidency'] = df['Age atend of presidency'].str.extract(r'(\d{1,2}\syears,\s\d{1,3}\sdays)')
df.head()
```

```python
df = df.rename(columns={'Age atend of presidency':'End of Presidency'})
df.head()
```

```python
df['End of Presidency']=df['End of Presidency'].str.extract(r'(?:^.*days)(\w{3}\s\d{1,2},\s\d{4})')
df.head()
```

```python
df['End of Presidency']=pd.to_datetime(df['End of Presidency'])
```

```python
df.head()
```

```python
df = df.rename(columns={'Post-presidencytimespan':'Years lived after Presidency'})
df.head()
```

```python
df['Died']= df['Died'].str.extract(r'(^\w{3}\s\d{1,2},\s\d{4})')
```

```python
df['Died'] = pd.to_datetime(df['Died'])
df.head()
```

```python
df
```

```python
df['Next_President_Start'] = df['Start of Presidency'].shift(-1)
```

```python
df.head()
```

```python
df['End of Presidency'] = df['End of Presidency'].fillna(df['Next_President_Start'])
df
```

```python
del(df['Next_President_Start'])
df
```

```python
columns_to_keep = ['President','Firstname','Lastname','Born','Start of Presidency','Age at the start of Presidency','End of Presidency','Age at the end of Presidency','Years lived after Presidency','Died','Age']
df = df[columns_to_keep]
df
```
